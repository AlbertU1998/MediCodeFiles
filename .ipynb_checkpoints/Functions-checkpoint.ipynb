def summarise_HR(HR_series):
    max_HR = max(HR_series)
    print("The maximum HR was", max_HR)
    
    min_HR = min(HR_series)
    print("The minimum HR was " + str(min_HR))

def summary(first, last, age, sex, weight = None, height = None, city = None):

        Now onto the principle of recursion. For people just getting into to programming this can be quite a daunting topic but really its a simple as calling a function within its own definition. Recursive implementations of various functions may be easily produced:
    </p>
</div>

<div class = "buffer"> </div>

```python
#determine the length of a container imperatively
def length(container):
    count = 0
    for thing in container:
        count = count + 1
    return count

#determine the length of a container recursively
def length(container):
    if not container: return 0
    return 1 + length(container[1:])
```

<div class = "back">
    <p class = "content">        
        Why would anyone want to do this?" I hear you scream with agony. Well the simple fact is that recursion <b>sometimes</b> allows us to simplify our programs. The best way to illustrate this is to use some examples. One of my favourite uss for recursion is in sorting. Typical sorting algorithims such as repeatedly scannig a list and removing the maximum value are quite slow (ie they require a high number of computations and comparisons than is optimal. Merge sort, implemented recursively below is much faster:
    </p>
</div>

<div class = "buffer"> </div>

```python
def merge_sort(container):
    if len(container) <= 1: return container
    
    #sort left and right halves
    midpoint = len(container) // 2
    left = merge_sort( container[:midpoint] )
    right = merge_sort( container[midpoint:] )
    
    #merge left and right halves
    out = []
    while left and right:
        if left[0] > right[0]:
            out += [ left[0] ]
            left = left[1:]
        else:
            out += [ right[0] ]
            right = right[1:]
            
    return out + (left if left else right)
```
<div class = "back">
    <p class = "content"> 
        Now you probably will not notice the speed enhancements untill you try the algorithms on larger data sets (by the way, the term algorithm and computer program are essentially interchangeable in this context). However they key here is that the function would be a lot less comprehensible if we had't used recursion!

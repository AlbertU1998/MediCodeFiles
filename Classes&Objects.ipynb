````python
#------------------------- your code here ---------------------------#
# Recursive class
class Recursive:
    def __init__(self):
        pass
    
    def length(self, container):
        if len(container) == 0:
            return 0
        else: 
            return 1 + self.length(container[:-1])
        
    def factorial(self, number):
        if number == 0: return 1
        return number * self.factorial(number - 1)
    
    def fancy_print(self, func, arg):
        if func == "factorial":
            print(arg, "! =", self.factorial(arg))
        elif func == "length":
            print(arg,"is", self.length(arg), "elements long")

re = Recursive()
re.fancy_print("length", [1,2,3])
re.fancy_print("factorial", 9)
````

#terser implmentation of merge sort
def merge_sort(container):
    if len(container) <= 1: return container

    #sort left and right halves
    midpoint = len(container) // 2
    left = merge_sort( container[:midpoint] )
    right = merge_sort( container[midpoint:] )

    #merge left and right halves
    out = []
    while left and right: 
        out.append( left.pop(0) if left[0] > right[0] else right.pop(0) )

    return out + (left if left else right)

print(merge_sort(my_list))

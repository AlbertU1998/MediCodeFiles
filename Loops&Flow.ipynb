```python
def reverso(a_list):
```

        The next type of loop we'll cover is <i>while</i> loops, these loops run for ever until a certain condition becomes true (remember our boolean type?) here are some examples:

<div class = "buffer"> </div>

```python
# declare a variable and add to it until it's greater than 19
k = 11
while k < 20:
    k = k + 1
    print(k)

#function that determines factorials: n! = n * (n - 1) * (n - 2) * ... * 1
def factorial(n):
    output = n
    while n > 1:
        n = n - 1
        output = output * n
    return output
         
# 5! = 120
print( factorial(5) )
```

    <p class = "content">
        These are great right? But what's our safe word? Python offers us a way to more tightly regulate the execution of for and while loops. The <b>continue</b> keyword lets you skip over an itertion of the for- or while- loop currently executing, try the examples below!
    </p>
</div>

<div class = "buffer"> </div>

```python
#get rid of repeats
names = ["Tom", "Ben", "Tom", "Clifford", "Jerry", "Daniel", "Daniel", "Ben"]
no_repeats = []
for name in names:
    if name in no_repeats:
        continue
    else:
        no_repeats = no_repeats + [name]
print(no_repeats)

#print odd numbers up to twenty one
c = 0
while c <= 20:
    c = c + 1
    if c % 2 == 0:
        continue
    print(c)
```

```python
i = 0

#detect dangerous changes in HR
hrs = [55, 57, 61, 77, 88, 89, 91, 95, 97, 100, 124, 134, 190]
before = 50
for hr in hrs:
    if hr - before >= 20:

        break
    before = hr
```

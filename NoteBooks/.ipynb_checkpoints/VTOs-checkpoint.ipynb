<div class="back">
    <p class = "content">
        One final type that will become particularly important later on is the boolean or 'bool'. Named after George Boole a pioneer in mathematic logic; this type can take but two values: True and False. Python also considers elements of other data types to be equivalent to boolean values: 0, empty lists and empty strings are all False!:
    </p>
</div>

<div class = "buffer"> </div>

```python
#all of these return False
False
bool(0)
bool("")
bool([])

#all of these return True
True
bool(1)
bool("this is true")
bool(["stuff", "in", "list"])
```

<div class="back">
    <p class = "content">
        Notice the use of the bool( ) function; we'll cover functions a ittle later but "type" functions like int( ), str( ), float( ), list( ) and tuple( ) simply convert things into that data type. This brings us onto our final group of operators, the comparators. These inlude your typical mathematical comparators such as "equals", "greater than" and "less than" represented with "==", "&gt;" and "&lt;" respectively. "Greater than or equal to" '&gt;=' and "less than or equal to" '&lt;=' are also available:
    </p>
</div>

<div class = "buffer"> </div>

```python
#all of these return True
42 == 42
42 > 21
42 < 96
42 <= 42
```

<div class="back">
    <p class = "content">
        Of course these may be used to compare non numeric data; strings are compared lexographically and lists are compared "pseudo" lexographically (first elements are compared, if equivalent then next elements compared):
    </p>
</div>

<div class = "buffer"> </div>

```python
#Aardvark comes before Bucket in the english dictionary so returns True
"Aardvark" < "Bucket"

#"pseudo" lexographically (pseudo because they're not words) returns True
["A", "a", "r", "d", "v", "a", "r", "k"] < ["B", "u", "c", "k", "e", "t"]

```

<div class="back">
    <p class = "content">
        An interesting point here is that "A" is less than "a", this has to do with the ASCII text representation. This is a widely used system that represents 256 different characters (letters numbers and symbols) each as 8-bit numbers. "A" is ASCII number 65 and "a" is 97. The ord( ) function wil reveal this to you. 
    </p>

<div class="back">
    <p class = "content">
        Finally, we also have operators that perform logical operations such as AND, OR, XOR and NOT. Their usage is exemplified below:
    </p>
</div>

<div class = "buffer"> </div>

```python
#AND returns True if both expressions are true:
bool("Tenzin" and 11)
bool(1 and 1)

#OR returns True if at least one expression is True:
bool(1 or 0)
bool("James" or [1,2,3])

#XOR returns true is exactly one of the expressions is True:
bool(1 ^ 0)
bool(0 ^ 56)

#NOT inverts a bool (return True is expression is False):
not False
not []
not 0

#all the below are False
bool([] and 0)
bool(0 or False)
bool(True ^ True)
not True
```
